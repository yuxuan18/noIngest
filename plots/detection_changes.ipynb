```python
import matplotlib.pyplot as plt
import pandas as pd
def plot_num_detects(filepath: str):
    df = pd.read_csv(filepath)
    data = df.groupby("frame_id").agg("count")
    data = data.sort_values(by=["frame_id"])["detect_id"]
    x = [i for i in range(len(data))]
    plt.figure(figsize=(3, 2))
    plt.plot(x, data)
    plt.xlabel("Frame ID")
    plt.ylabel("# objects")
    video_id = filepath.split("/")[-1].split(".")[0]
    plt.savefig(f"figures/detects_{video_id}.pdf", bbox_inches='tight')
```

```python
plot_num_detects("../data/vehicle/detects/41.csv")
```
